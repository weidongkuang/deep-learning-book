```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
# Fixing random state for reproducibility
np.random.seed(1968)
x=np.random.rand(10)  # uniform [0,1)
y=4+6*x+np.random.randn(10) # randn: Normal with mean=0, var=1
```

```python
plt.scatter(x, y, alpha=0.9)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# normal equation
x_bar=statistics.mean(x)
y_bar=statistics.mean(y)
theta1=sum(np.multiply((x-x_bar),(y-y_bar)))/sum(np.multiply((x-x_bar),(x-x_bar)))
theta0=y_bar-theta1*x_bar
```

```python
x_new=np.array([0,1])
y_predict=theta1*x_new+theta0
y_true=6*x_new+4
plt.plot(x_new,y_predict,"r-", label='linear regression line')
plt.plot(x_new,y_true,"b--", label='target linear line')
plt.plot(x,y,"b.")
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.legend(loc='lower right')
plt.xlim([0,1])
plt.ylim([0,12])
plt.show()
```

```python
# plot regression lines for different datasets
theta1s=[]
theta0s=[]
for i in range (0,5):
    np.random.seed(i)
    x=np.random.rand(10)  # uniform [0,1)
    y=4+6*x+np.random.randn(10) # randn: Normal with mean=0, var=1
    
    # normal equation
    x_bar=statistics.mean(x)
    y_bar=statistics.mean(y)
    theta1=sum(np.multiply((x-x_bar),(y-y_bar)))/sum(np.multiply((x-x_bar),(x-x_bar)))
    theta0=y_bar-theta1*x_bar
    theta1s.append(theta1)
    theta0s.append(theta0)
x_new=np.array([0,1])
y_true=6*x_new+4

y_predict0=theta1s[0]*x_new+theta0s[0]
y_predict1=theta1s[1]*x_new+theta0s[1]
y_predict2=theta1s[2]*x_new+theta0s[2]
y_predict3=theta1s[3]*x_new+theta0s[3]
y_predict4=theta1s[4]*x_new+theta0s[4]

plt.plot(x_new,y_predict0,"g-", label='linear regression line1')
plt.plot(x_new,y_predict1,"c-", label='linear regression line2')
plt.plot(x_new,y_predict2,"m-", label='linear regression line3')
plt.plot(x_new,y_predict3,"y-", label='linear regression line4')
plt.plot(x_new,y_predict4,"k-", label='linear regression line5')

plt.plot(x_new,y_true,"b--", label='target linear line', linewidth=3)
#plt.plot(x,y,"b.")
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.legend(loc='lower right')
plt.xlim([0,1])
plt.ylim([0,12])
plt.show()
```

```python
# gradient descent algorithm for a single feature input

# Fixing random state for reproducibility
np.random.seed(1968)
x=np.random.rand(10)
y=4+6*x+np.random.randn(10)

# gradient descent algorithm
theta0=1
theta1=1
alpha=1
loss=[]
m=x.shape[0]

for iter in range(0, 100):
    error=theta0+theta1*x-y
    temp0=theta0-alpha*statistics.mean(error)
    temp1=theta1-alpha*statistics.mean(error*x)
    theta0=temp0
    theta1=temp1
    J=sum((theta0+theta1*x-y)**2)/(2*m)
    loss.append(J)
```

```python
# plot the linear model and training examples
x_new=np.array([0,1])
y_predict=theta1*x_new+theta0
plt.plot(x_new,y_predict,"r-")
plt.plot(x,y,"b.")
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.xlim([0,1])
plt.ylim([0,12])
plt.show()
```

```python
# plot the loss function vs iteration steps
plt.plot(loss, 'b-')
plt.xlabel('iteration', fontsize="18")
plt.ylabel('loss', fontsize="18")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
#linear model with two features
# generate training examples, x(m,3), y(m,1)
np.random.seed(1969) # for reproducibility
m=10
x0=np.ones(m).reshape((m,1))
x=np.random.rand(m,2)
x=np.concatenate((x0,x),axis=1)
w=np.array([4,6,-1])
y=np.dot(x,w)+0.5*np.random.randn(10)
```

```python
#gradient descent algorithm for multiple features
loss=[]
n=x.shape[1]
theta=np.ones(n) 
# initial values for parameters, theta=np.array([1,1,1])

alpha=0.1 #learning rate
for iter in range(0, 1000):
    hypothesis=np.dot(x, theta)
    error=hypothesis-y
    J=np.sum(error**2)/(2*m)
    gradient=np.dot(np.transpose(x), error)/m
    theta=theta-alpha*gradient    
    loss.append(J)
```

```python
# plot the loss function vs iteration steps
plt.plot(loss, 'b-')
plt.xlabel('iteration', fontsize="18")
plt.ylabel('loss', fontsize="18")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
print('true theta:',w)
print('predicted theta:',theta)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
#num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'furnishingstatus']

#df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
X_train_scaled = scaler.fit_transform(X_train) #train is numpy array (381,12)
feature_min=scaler.data_min_   # shape (12.)
feature_max=scaler.data_max_   # shape (12,)
feature_range=scaler.data_range_  # shape (12,)
```

```python
X_test_scaled=(X_test-feature_min)/feature_range
```

```python
def gradient_descent(alpha, x, y, numIterations):  
    cost=[] 
    m = x.shape[0] # number of samples  
    x0=np.ones(m).reshape((m,1))  
    x=np.concatenate((x0,x), axis=1)  
    theta = np.ones(x.shape[1]) 
    x_transpose = x.transpose()  
    for iter in range(0, numIterations):  
        hypothesis = np.dot(x, theta)  
        loss = hypothesis - y  
        J = np.sum(loss ** 2) / (2 * m)  # cost  
        #print ("iter %s | J: %.3f" % (iter, J))        
        gradient = np.dot(x_transpose, loss) / m           
        theta = theta - alpha * gradient  # update  
        cost.append(J)
        #theta_list.append(theta)
    return theta,cost
```

```python
theta ,cost= gradient_descent(alpha=0.01, x=X_train_scaled, y=y_train, numIterations=10000)  
```

```python
def predict(x, theta):  
    m = x.shape[0] # number of samples  
    x0=np.ones(m).reshape((m,1))  
    x=np.concatenate((x0,x), axis=1) # add the first column with "1"  
    return np.dot(x,theta)  
```

```python
Y_test_pred=predict(X_test_scaled,theta)  
```
